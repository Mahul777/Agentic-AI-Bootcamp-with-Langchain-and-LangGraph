### Orchestrator-Worker

In the orchestrator-workers workflow, a central LLM dynamically breaks down tasks, delegates them to worker LLMs, and synthesizes their results.

**When to use this workflow:** This workflow is well-suited for complex tasks where you can't predict the subtasks needed (in coding, for example, the number of files that need to be changed and the nature of the change in each file likely depend on the task). Whereas it's topographically similar, the key difference from parallelization is its flexibility—subtasks aren't pre-defined, but determined by the orchestrator based on the specific input.

```python
#📦 1. Import and Environment Setup
import os
from dotenv import load_dotenv
# Import Send if not already imported
from langgraph.graph import StateGraph, START, END
from langgraph.types import Send


load_dotenv()
#•	Loads environment variables (like API keys) from a .env file.

#🔐 2. Set the API Key for Groq
from langchain_groq import ChatGroq
os.environ["GROQ_API_KEY"] = os.getenv("GROQ_API_KEY")
#•	Sets the API key so you can use the Qwen LLM from Groq.

#🧠 3. Initialize the LLM
from langchain_groq import ChatGroq
llm = ChatGroq(model="llama-3.3-70b-versatile")
result = llm.invoke("Hello")
result
# AIMessage(content="Hello. It's nice to meet you. Is there something I can help you with or would you like to chat?", additional_kwargs={}, response_metadata={'token_usage': {'completion_tokens': 25, 'prompt_tokens': 36, 'total_tokens': 61, 'completion_time': 0.030250964, 'prompt_time': 0.010383945, 'queue_time': 0.044443415, 'total_time': 0.040634909}, 'model_name': 'llama-3.3-70b-versatile', 'system_fingerprint': 'fp_3f3b593e33', 'service_tier': 'on_demand', 'finish_reason': 'stop', 'logprobs': None}, id='run--ff774499-1e65-4aa8-962c-bb01f74fbc46-0', usage_metadata={'input_tokens': 36, 'output_tokens': 25, 'total_tokens': 61})
#•	You set up and test the LLM with a simple “Hello” message.

# 📋 4. Define Planning Schema
from pydantic import BaseModel
from typing import List
class Section(BaseModel):
    name: str
    description: str

class Sections(BaseModel):
    sections: List[Section]

#     •	Tells the LLM that when it plans a report, it should return:
# o	A list of sections.
# o	Each section has a name and description.
#📍 5. Planner
planner = llm.with_structured_output(Sections)
#•	This wraps the LLM so it outputs results following the schema above.

#🧠 6. Define State for LangGraph
from typing import TypedDict
from typing_extensions import Annotated
import operator
class State(TypedDict):
    topic: str
    sections: list[Section]
    completed_sections: Annotated[list, operator.add]
    final_report: str
# •	This defines the data structure passed between steps.
# o	topic: The report topic.
# o	sections: List of planned sections.
# o	completed_sections: Filled in by multiple LLM workers in parallel.
# o	final_report: Final combined output.

#🧱 7. Worker State
class WorkerState(TypedDict):
    section: Section
    completed_sections: Annotated[list, operator.add]
#•	This defines the input/output format for a worker who writes one section.

#⚙️ 8. Define Nodes (Steps)
from langchain.schema import SystemMessage, HumanMessage
#🧑‍✈️ a) Orchestrator: Makes the report plan
def orchestrator(state: State):
    report_sections = planner.invoke([
        SystemMessage(content="Generate a plan for the report."),
        HumanMessage(content=f"Here is the report topic: {state['topic']}")
    ])
    return {"sections": report_sections.sections}
#•	Uses the planner to generate the structure of the report.
#•	Example: 3 sections like Introduction, Use Cases, Future.
#✍️ b) Worker: Writes each section
def llm_call(state: WorkerState):
    section = llm.invoke([
        SystemMessage(content="Write a report section..."),
        HumanMessage(content=f"Here is the section name: {state['section'].name} and description: {state['section'].description}")
    ])
    return {"completed_sections": [section.content]}
#•	Each worker writes one section using the name and description.
#👥 c) Assign Workers
def assign_workers(state: State):
    return [Send("llm_call", {"section": s}) for s in state["sections"]]
#•	This function assigns one worker per section.
#•	All workers run in parallel using Send().
#🧵 d) Synthesizer: Combine all sections
def synthesizer(state: State):
    completed_sections = state["completed_sections"]
    completed_report_sections = "\n\n---\n\n".join(completed_sections)
    return {"final_report": completed_report_sections}
#•	Joins all the written sections into one complete report, separated by lines.
#🧩 9. Build the Graph
from langgraph.graph import StateGraph, START, END
orchestrator_worker_builder = StateGraph(State)
#•	Creates a new LangGraph workflow, which is like a flowchart of AI steps.

#🧱 Add Nodes
orchestrator_worker_builder.add_node("orchestrator", orchestrator)
orchestrator_worker_builder.add_node("llm_call", llm_call)
orchestrator_worker_builder.add_node("synthesizer", synthesizer)
#🔗 Connect Nodes
orchestrator_worker_builder.add_edge(START, "orchestrator")
orchestrator_worker_builder.add_conditional_edges(
    "orchestrator", assign_workers, ["llm_call"]
)
orchestrator_worker_builder.add_edge("llm_call", "synthesizer")
orchestrator_worker_builder.add_edge("synthesizer", END)
# Flow:
# 1.	Start → Orchestrator (plan)
# 2.	Orchestrator → Many Workers (write sections)
# 3.	All Workers → Synthesizer (combine sections)
# 4.	Synthesizer → End

#🛠️ 10. Compile the Graph
from IPython.display import Image, display
orchestrator_worker = orchestrator_worker_builder.compile()
#•	Finalizes the graph into an executable workflow.
#🧾 11. Visualize the Workflow
display(Image(orchestrator_worker.get_graph().draw_mermaid_png()))
#•	Optional: Show a diagram of the workflow.

#🚀 12. Run the Workflow

state = orchestrator_worker.invoke({"topic": "Create a report on Agentic AI RAGs"})
# •	You ask the system to create a full report on Agentic AI RAGs.
# •	It:
# 1.	Plans sections
# 2.	Writes each one
# 3.	Combines them
# 🖨️ 13. Display the Report
from IPython.display import Markdown
Markdown(state["final_report"])
# •	Shows the final report in markdown format, nicely formatted.
```
